# Removing Unnecessary Hierarchies

For our project, we are interested in Hierarchy values 3, 5, 6, and 7.<br>
3: Robbery: implies there is force or threat of force involved <br>
5: Burglary: trespassing involved <br>
6: Theft: no threat of force or burglary <br>
7: Vehicle Theft: same as theft <br>

So I'll drop the useless Hierarchy values to remove some clutter.

```python
data = pd.read_csv("archive-police-blotter.csv", parse_dates=True, low_memory=False)
data.drop(columns = ["PK", "CCR", "INCIDENTTIME", "INCIDENTLOCATION", "CLEAREDFLAG", "INCIDENTZONE", "HIERARCHYDESC", "INCIDENTTRACT", "X", "Y"], inplace = True)

#Deletes from 8-29
for i in range(22):
    data.drop(data[data.HIERARCHY == i+8].index, inplace=True)

#Deletes the ones the if statement didn't get to
data.drop(data[data.HIERARCHY == 0].index, inplace=True)
data.drop(data[data.HIERARCHY == 1].index, inplace=True)
data.drop(data[data.HIERARCHY == 2].index, inplace=True)
data.drop(data[data.HIERARCHY == 4].index, inplace=True)
data.drop(data[data.HIERARCHY == 99].index, inplace=True)

data
```

# Data Analysis

I'd like to first explain the metric I'll use to grade the neighborhoods. First, I'm going to assume there are two types of criminals: those that are confident in their skills, and those that are relatively new and inexperienced. <br>

For the first type of criminal, I'm going to assume that the best neighborhoods to rob are those with a low crime rate; this means people won't be expecting crime as much. Obviously, there is the possibility that crime is low because the police are really good or because security is tight, but I am going to assume that a good criminal will not care or will be able to bypass these obstacles with skills. </br>

For the second type of criminal, I'm going to assume they're inexperienced, and would like to commit crime in a neighborhood that is easier to succeed without being caught. This would be a neighborhood with a high crime rate. </br>

# Conclusions

For the  high-class criminal (the first type), the best neighborhoods to rob would be Mt. Oliver Borough, Chartiers City, and New Homestead, all of which have less than 200 offenses. </br>

For the inferior criminal (the second type), the best neighborhoods to rob would be Golden Triangle/Civic Arena, South Side Flats, Shadyside, and Bloomfield, all of which have more than 6000 offenses.

# Possible Errors

There are a couple of ways this data could be incorrect, which is why it's important to combine it with other datasets and information for a more accurate picture. Setting that aside, I'll take a moment to list some of the faults of this analysis specifically. </br>

This set does not take size or population into account; neighborhoods with a larger size & population will naturally have more crime, while smaller neighborhoods will have less. This means that some of the neighborhoods with the highest amount of offenses might just be very large neighborhoods, and vice versa for smaller ones. </br>

This set also does not look into how effective the police are at catching criminals. It's possible one of the cities with lots of crime also has an incredibly competent police force with a high rate of success.
